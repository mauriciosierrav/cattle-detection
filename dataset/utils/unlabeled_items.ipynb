```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_ = cv2.imread('../images/val/5df085b3-G2072432.JPG')
img = cv2.cvtColor(img_, cv2.COLOR_BGR2RGB)  #Convertir de BGR a RGB

dh, dw, _ = img.shape

fl = open('../labels/val/5df085b3-G2072432.txt', 'r')
data = fl.readlines()
fl.close()

for dt in data:
    _, x, y, w, h = map(float, dt.split(' '))
    l, r, t, b = int((x - w / 2) * dw), int((x + w / 2) * dw), int((y - h / 2) * dh), int((y + h / 2) * dh)
    l, r, t, b = max(0, l), min(dw - 1, r), max(0, t), min(dh - 1, b)
    cv2.rectangle(img, (l, t), (r, b), (255, 0, 0), 10)

plt.imshow(img)
plt.axis('off')
plt.show()

# Guardar la imagen con los recuadros dibujados
_img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)  #Convertir de BGR a RGB
cv2.imwrite('/Users/jmsierra/muestra.jpg', _img)
```
