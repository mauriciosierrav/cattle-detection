```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt

image = Image.open("./sample_image.jpg")
mirrored_image = image.transpose(Image.FLIP_LEFT_RIGHT)
rotated_image = image.rotate(45)
brightness_enhancer = ImageEnhance.Brightness(image)
brighter_image = brightness_enhancer.enhance(factor=2)

imagenes = [image, mirrored_image, rotated_image, brighter_image]
titulos = ['Original', 'Mirroed', 'Rotated', 'Brighter']

fig, axs = plt.subplots(1, 4, figsize=(10, 8))
axs = axs.flatten()
for i, ax in enumerate(axs):
    ax.imshow(imagenes[i])
    ax.set_title(titulos[i])
    ax.axis('off')

plt.tight_layout()  # Ajustar los espacios entre subplots
plt.show()
```
